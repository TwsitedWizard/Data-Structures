### Bellman Ford

```python
class Edge:
    def __init__(self, u, v, weight):
        self.u = u
        self.v = v
        self.weight = weight

def bellman_ford(graph, V, source):
    # Initialize time
    time = 0
    
    # Step 1: Initialize distances from source to all vertices
    distances = [float('inf')] * V
    distances[source] = 0
    time += 1  # Initialization step takes 1 unit of time

    # Step 2: Relax all edges V-1 times
    for _ in range(V - 1):
        for edge in graph:
            if distances[edge.u] != float('inf') and distances[edge.u] + edge.weight < distances[edge.v]:
                distances[edge.v] = distances[edge.u] + edge.weight
            time += 1  # Each relaxation step takes 1 unit of time

    # Step 3: Check for negative weight cycle
    for edge in graph:
        if distances[edge.u] != float('inf') and distances[edge.u] + edge.weight < distances[edge.v]:
            print("Graph contains negative weight cycle")
            return None, time  # Negative cycle detected

    return distances, time

if __name__ == "__main__":
    # Number of vertices
    V = 6
    # List of edges (u, v, weight)
    edges = [
        Edge(0, 1, 1),
        Edge(0, 2, 4),
        Edge(1, 2, 3),
        Edge(1, 3, 2),
        Edge(1, 4, 2),
        Edge(3, 2, 5),
        Edge(3, 1, 1),
        Edge(4, 3, 3),
        Edge(4, 5, 6),
        Edge(3, 5, 3)
    ]

    source = 0 

    # Call Bellman-Ford algorithm
    distances, time = bellman_ford(edges, V, source)

    if distances:
        print(f"Shortest distances from source vertex {source}:")
        for i in range(V):
            print(f"Vertex {i}: {distances[i]}")

    print(f"Total time taken: {time} units")
```

```python
def matrix_chain_order(p):
    time = 0  # Initialize time
    n = len(p) - 1  # Number of matrices
    # m[i, j] will store the minimum cost of multiplying matrices Ai..Aj
    m = [[0 for _ in range(n)] for _ in range(n)]
    # s[i, j] will store the index of the matrix after which the multiplication is split
    s = [[0 for _ in range(n)] for _ in range(n)]

    # Compute the minimum cost for chains of increasing length
    for chain_length in range(2, n+1):  # chain_length ranges from 2 to n
        for i in range(n - chain_length + 1):
            j = i + chain_length - 1
            m[i][j] = float('inf')  # Initialize the cost to a large value
            for k in range(i, j):
                # q is the cost for multiplying matrices Ai..Ak and A(k+1)..Aj
                q = m[i][k] + m[k+1][j] + p[i] * p[k+1] * p[j+1]
                time += 1  # Increment time for each comparison and calculation
                if q < m[i][j]:
                    m[i][j] = q
                    s[i][j] = k  # Store the split index
                    time += 1  # Increment time for each update of m[i][j]

    # Return the minimum cost, the matrix split information, and the time taken
    return m, s, time

# Function to print the optimal parenthesization
def print_optimal_parenthesization(s, i, j):
    if i == j:
        print(f"A{i+1}", end="")
    else:
        print("(", end="")
        print_optimal_parenthesization(s, i, s[i][j])
        print(" x ", end="")
        print_optimal_parenthesization(s, s[i][j] + 1, j)
        print(")", end="")

# Example usage
if __name__ == "__main__":
    # Dimensions of the matrices A1, A2, ..., A4
    p = [30, 35, 15, 5, 10, 20, 25]

    # Call matrix_chain_order to get the minimum cost and the optimal split points
    m, s, time = matrix_chain_order(p)

    # The minimum cost of multiplying the matrices
    print(f"Minimum number of scalar multiplications: {m[0][len(p) - 2]}")

    # Print the optimal parenthesization
    print("Optimal parenthesization is: ", end="")
    print_optimal_parenthesization(s, 0, len(p) - 2)
    print()  # Newline for better formatting

    # Print total time taken (in terms of steps)
    print(f"Total time taken: {time} units")
```

### Kruskals Algorithm

```python
class DisjointSet:
    def __init__(self, n):
        self.parent = list(range(n))  # Each vertex is its own parent initially
        self.rank = [0] * n  # Rank is used for optimization (union by rank)

    # Find with path compression
    def find(self, u):
        if self.parent[u] != u:
            self.parent[u] = self.find(self.parent[u])  # Path compression
        return self.parent[u]

    # Union by rank
    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:  # If u and v are not in the same set
            # Union by rank optimization: attach the smaller tree under the larger tree
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

class Kruskal:
    def __init__(self, vertices, edges):
        self.V = vertices  # Number of vertices
        self.edges = edges  # List of edges in the form (u, v, weight)

    # Kruskal's Algorithm to find MST
    def kruskal_mst(self):
        mst = []  # Result to store MST
        disjoint_set = DisjointSet(self.V)

        # Step 1: Sort all edges in non-decreasing order of their weight
        self.edges.sort(key=lambda x: x[2])

        # Step 2: Process edges one by one
        for u, v, weight in self.edges:
            # Step 2a: Find the roots of the sets to which u and v belong
            root_u = disjoint_set.find(u)
            root_v = disjoint_set.find(v)

            # Step 2b: If they are in different sets, include this edge in MST
            if root_u != root_v:
                mst.append((u, v, weight))  # Add edge to MST
                disjoint_set.union(u, v)  # Union the sets containing u and v

        return mst

# Example Usage
if __name__ == "__main__":
    # Number of vertices
    V = 4
    # List of edges in the format (u, v, weight) where u and v are vertices and weight is the edge weight
    edges = [
        (0, 1, 10),
        (0, 2, 6),
        (0, 3, 5),
        (1, 3, 15),
        (2, 3, 4)
    ]

    # Initialize Kruskal's algorithm with the number of vertices and the edges
    kruskal = Kruskal(V, edges)

    # Call kruskal_mst to get the Minimum Spanning Tree
    mst = kruskal.kruskal_mst()

    # Output the MST
    print("Edges in the Minimum Spanning Tree (MST):")
    for u, v, weight in mst:
        print(f"({u}, {v}) -> {weight}")
```

```python
class DisjointSet:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, u):
        if self.parent[u] != u:
            self.parent[u] = self.find(self.parent[u])
        return self.parent[u]

    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

class Kruskal:
    def __init__(self, vertices, edges):
        self.V = vertices
        self.edges = edges

    def kruskal_mst(self):
        mst = []
        disjoint_set = DisjointSet(self.V)
        time = 0  # Initialize time count

        # Step 1: Sort edges by weight
        self.edges.sort(key=lambda x: x[2])
        time += len(self.edges) * 2  # Approximate sorting time based on edge count

        # Step 2: Process each edge
        for u, v, weight in self.edges:
            root_u = disjoint_set.find(u)
            root_v = disjoint_set.find(v)
            time += 2  # Counting the find operations

            if root_u != root_v:
                mst.append((u, v, weight))
                disjoint_set.union(u, v)
                time += 1  # Counting union operation

        return mst, time

# Posteriori analysis function
def posteriori_analysis_kruskal():
    vertex_counts = [10, 20, 30, 40]  # Various numbers of vertices for analysis
    edges_per_vertex = 5  # Average number of edges per vertex
    times = []

    for V in vertex_counts:
        edges = []
        for i in range(V):
            for j in range(i + 1, V):
                weight = random.randint(1, 100)
                edges.append((i, j, weight))
                if len(edges) >= V * edges_per_vertex:
                    break

        # Initialize and run Kruskal's algorithm
        kruskal = Kruskal(V, edges)
        _, time_taken = kruskal.kruskal_mst()
        times.append(time_taken)

    # Plotting results
    plt.figure(figsize=(10, 6))
    plt.plot(vertex_counts, times, marker='o', color='b', label="Kruskal's Algorithm Time Complexity")
    plt.xlabel("Number of Vertices")
    plt.ylabel("Time Taken (Simulated Time Steps)")
    plt.title("Posteriori Analysis of Kruskal's Algorithm Time Complexity")
    plt.legend()
    plt.grid(True)
    plt.show()

# Execute posteriori analysis
posteriori_analysis_kruskal()
```
